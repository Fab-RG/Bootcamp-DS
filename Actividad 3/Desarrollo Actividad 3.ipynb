- **¿Aumentó el registro de vehículos eléctricos durante los últimos cuatro años?**

- **¿En qué año se registró la mayor cantidad de vehículos eléctricos?**

```python
cantidad = Dataset['Model Year'].value_counts().head()
cantidad
```

```python
x = Dataset['Model Year'].value_counts().head().index
y = Dataset['Model Year'].value_counts().head().values 
sns.barplot(x=x, y=y)
plt.xticks(rotation=45)
plt.xlabel('Año')
plt.ylabel('Cantidad de Vehículos')
plt.title('Cantidad de Vehículos Registrados')
```

- **Respuesta primer pregunta**: Se evidencia que durante los últimos cuatro años, aumentó el registro de vehículos eléctricos cada año, respecto al anterior.

- **Respuesta segunda pregunta**: En el año 2023 se evidenció la mayor cantidad de registro de vehículos eléctricos, aumentando más del doble de la cantidad registrada el año anterior.

- **¿Cuál es la marca de vehículos eléctricos que más se registró en los últimos cuatro años?**

```python
gr2 = Dataset[['Make','Model Year']].value_counts().head(10)
gr2
```

```python
box = Dataset[['Make','Model Year']][0:10]
sns.boxplot(data=box, x = 'Model Year', y = 'Make', showfliers= False)
plt.xlabel('Año')
plt.ylabel('Marca')
plt.title('Boxplot de Marcas más Registradas')
```

**Respuesta:** Se evidencia que Tesla fue la marca de vehículos eléctricos que más se registró en los últimos cuatro años. 

- **¿Cuál es la ciudad donde se registraron la mayor cantidad de vehículos eléctricos?**

```python
Dataset['City'].value_counts().head(10)
```

```python
Dataset['City'].value_counts()[0:10].plot(kind='bar',figsize=(12,6))
plt.xlabel('Ciudad')
plt.ylabel('Cantidad de Vehículos Registrados')
plt.title('Diagrama de Vehículos Eléctricos en ciudades del Estado de Washington, EEUU')
```

**Respuesta:** Se evidencia que la ciudad de Seattle, es donde se han registrado la mayor cantidad de vehículos eléctricos del Estado de Washington, EEUU.
